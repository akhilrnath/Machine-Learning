```python
# import dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

1.	Read the iris dataset using Pandas (ref Option 2 above)
2.	Extract the Iris-setosa and Iris-verginica samples only
3.	Extract the petal length and petal width. These shall be our features.

```python

df = pd.read_csv('iris_data.csv', header=0)

# Read the data from the .csv, using the first row as the headers of the data
data = pd.read_csv('/content/iris_data.csv',header=0)

# Show just the first few samples so we know it's read correctly
print(data.head(n=300))
```

```python
#Extract the Iris-setosa and Iris-verginica samples only
y1 = data.iloc[0:150, 4].values
y = np.where(y1 == 'Iris-setosa', -1, np.where(y1 == 'Iris-virginica', 1, 0))
y = y[y != 0]

# Extract the rows containing Iris-setosa and Iris-verginica only
#X = data.iloc[0:150, [2,3]].values
petal= pd.concat([df.iloc[0:50], df.iloc[100:150]])

# Extract the petal length and petal width for the selected rows
X = petal.iloc[:, [2,3]].values


```

4.	Plot the dataset as a scatter plot with the two Iris types shown as different colours

```python
plt.scatter(X[y == -1][:, 0], X[y == -1][:, 1], label='Iris-setosa', c='b', marker='o')
plt.scatter(X[y == 1][:, 0], X[y == 1][:, 1], label='Iris-virginica', c='r', marker='x')
# Label the axes
plt.xlabel('Petal length')
plt.ylabel('Petal width')

# Add a legend
plt.legend(loc='upper left')

# Show the plot
plt.show()
```

5.	Use the perceptron in sklearn.linear_model to train on the Iris data previously used. Set verbose = 1; have a look at the epochs, it converges very quickly?

```python
# So as not to confuse with our previously defined Perceptron class
from sklearn.linear_model import Perceptron
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

ppn = Perceptron(verbose=1)
ppn.fit(X_train, y_train)
```

6.	Using many random inputs (e.g. 5000) plot to see if you can see the decision boundary for the feature scaled case

```python
import random

np.random.seed(0)

#Generating random petal length and petal width
new_petal_length =np.random.uniform(1,6,5000)
new_petal_width= np.random.uniform(0.1,2.4,5000)
X_data = np.column_stack((new_petal_width, new_petal_length))

#predict the data using trained perceptron
new_predict=ppn.predict(X_data)
X_data_predict = np.column_stack((X_data, new_predict))
#print(X_data_predict)

#plotting the data
plt.figure(figsize=(8, 6))
plt.scatter(X_data_predict[:, 1], X_data_predict[:, 0], c=X_data_predict[:, 2], cmap='coolwarm', label='Iris Virginica (Red),Iris Setosa(Blue)')

# Set labels and legend
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.legend(loc='upper left')
```

**Commonality and differences between artificial neurons and biological neurons**

Learning ability is considered one main aspect of intelligence. Like human brain ,machines has the ability to learn and extrapolate from data sets to perform some tedious tasks like classifying  with the help of artificial neuron network. Artificial neurons were inspired by the biological neurons of neuroscience and they have many similarities since it is designed to mimic the basic functions of biological neurons and also artificial neurons have many difference from their biological counter parts.

when information is processed, both works in similar manner . In biological neural network, processing abilities follow highly parallel processes operating on representations that are distributed over many neurons. In artificial neuron network also processing abilities  are captured in highly parallel computation based on distributed representations. The artificial neuron receives inputs analogous to the electrochemical impulses the dendrites of biological neurons receive from another neuron. Similar to the physical changes that occur in the synapses ,artificial signals can be changed by functions in a manner. The output of the artificial neuron corresponds to signals sent out from biological neuron over its axon. Biological neurons fire an action potential when their membrane potential exceeds a certain threshold, while artificial neurons activate (output a signal) when the weighted sum of inputs surpasses a predefined threshold.

Unlike its biological counter part, Artificial Neural Networks (ANNs) are trained from scratch, depending upon  problem and architecture at hand using a fixed topology. In case of fault tolerance,  due to their topology biological neuron networks are also fault-tolerant and also here the Information is stored redundantly which will not result in memory loss due to some minor failures. The human brain can also  heal and recover to an extent by itself  But Artificial neural networks are not suitably modeled for self-generation and fault tolerance. Eventhough we are advanced in all field ,we still do not understand how brains learn, or how it stores data and how it recall all the information and make connection between information. On the other hand Artificial neural networks , have a predefined model, where additional or removal of any neurons or their connection cannot be possible. But we can make amendments in model .

Artificial neural networks as of today  are still in their infancy. While analogous in structure, and created for mimicking biological neurons  with respect to notions of weights, neurons (functional units), topology, and learning algorithms, still they are not yet capable of doing what brain do considering the bigger-picture. Furthermore, they cannot be used to work on many heterogeneous tasks or trained to do complex tasks  simultaneously. But times change ,the Artificial neuron networks are used in almost all fields now and significant amount of studies are conducting and thus  ANNs will definitely  begin to perform more complex classes of tasks in near future even like , near-human and may be more than humans -superhuman levels, perhaps becoming  structurally and functionally  more similar to biological neural networks.
